```python
import numpy as np
from scipy import integrate
import scipy
import scipy.optimize
import scipy.integrate
import matplotlib.pyplot as plt
from autograd import numpy as anp
```

```python
n_data = 305
domain = np.arange(n_data)
# D = np.array([0, 0, 0])#10 * np.ones(n_data)
path = "/Users/benchristensen/Desktop/School/Winter 2019/Math 438/optimal_inventory/"
D = np.load(path+"d.npy")
w = 4
c = 6

def L(x, t):
    '''
    Lagrangian
    x (function): state
    t (value): time'''
    t = int(t)
    cost = max(c * (x[t] - D[t], w * (D[t] - x[t])))
    return cost

def solve_ode(f_prime, u, x0, domain):
    '''
    Solves an ODE
    f_prime (function): x' = f_prime
    x0 (value): inital x
    u (array): optimal control
    domain (array): domain
    '''
    x = np.zeros_like(domain)
    x[0] = x0
    for i in range(1, len(domain)):
        x[i] = x[i-1] + f_prime(u, i)
    return x

# def solve_ode(f_prime, u, x0, domain):
#     print("u:", u)
#     result = []
#     for i in range(len(domain)):
#         print(i)
#         function = f_prime(u, i)
#         print(function)
#         integrated = integrate.simps(function)
#         result.append(integrated)
#     result = np.array(result)
#     print(result)
#     constant = x0 - result[0]
#     return result + constant


def f_prime(u, t):
    '''x = f_prime
    u (array): optimal control
    t (value): time'''
    return u[t] - D[t]

# def f_prime(u, t):
#     return u[:t+1] - D[:t+1]

def J(u):
    '''Cost functional
    u (value): optimal control'''
    x = solve_ode(f_prime, u, 0, domain)
    cost = integrate.simps([L(x,t) for t in domain])
    return x, cost

def u_star():
    '''Finds the optimal control
    '''
    u0 = 600*np.ones(n_data)
    u = scipy.optimize.fmin(func=lambda u: J(u)[1], x0=u0)
    return u

u = np.zeros(n_data)
x, cost = J(u)
print("Cost: {:.3f}".format(cost))
plt.plot(x)
plt.show()

u = u_star()
```
